```python
import pandas as pd
```

```python
data = pd.read_csv('../data/clean_data.csv', index_col = 0)
data.head()
```

```python
daily_vaccinations = data[['country', 'date', 'daily_vaccinations', 'population']]
```

```python
daily_vaccinations.loc[:, 'daily_vac_per_capita'] = daily_vaccinations.loc[:, 'daily_vaccinations']/daily_vaccinations.loc[:, 'population']
daily_vaccinations.loc[:, 'date'] = pd.to_datetime(daily_vaccinations.loc[:, 'date'], format='%Y-%m-%d')
daily_vaccinations = daily_vaccinations[['country', 'date', 'daily_vac_per_capita']]
```

```python
import pandas as pd
import numpy as np
from bokeh.palettes import cividis
from bokeh.plotting import figure, show, output_file
```

```python
#countries = daily_vaccinations['country'].unique().tolist()
countries = ['United States', 'United Kingdom', 'Israel', 'France', 'China']
```

```python
p = figure(x_axis_type='datetime', plot_width = 800, plot_height = 800, tools = 'hover')

cividis_colors = cividis(len(countries))
for i in range(len(countries)):
    data = daily_vaccinations[daily_vaccinations['country'] == countries[i]]
    p.line(x='date', y='daily_vac_per_capita', line_width=2, source=data, legend_label=countries[i], color = cividis_colors[i])

p.legend.location = "top_left"
p.legend.click_policy="hide"

show(p)
```

Inputs: Date slider, multi-choice countries, daily vaccination range (slider)
