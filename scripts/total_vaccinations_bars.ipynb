```python
import pandas as pd
```

```python
data = pd.read_csv('../data/clean_data.csv', index_col = 0)
data.head()
```

```python
total_vaccinations = data[['country', 'iso_code', 'date', 'total_vaccinations', 'population']]
```

```python
from bokeh.io import output_notebook, show, output_file
from bokeh.plotting import figure
from bokeh.models import GeoJSONDataSource, LinearColorMapper, ColorBar
from bokeh.palettes import brewer
```

## Total Vaccinations per Capita (Bar Graph)

```python
total_vaccinations.loc[:, 'total_vaccinations_per_capita'] = total_vaccinations.loc[:, 'total_vaccinations']/total_vaccinations.loc[:,'population']
total_vaccinations.loc[:, 'date'] = pd.to_datetime(total_vaccinations.loc[:, 'date'], format='%Y-%m-%d')
max_total_vaccinations = total_vaccinations.groupby('country')['total_vaccinations_per_capita'].max().sort_values()

countries = max_total_vaccinations.index.to_list()
counts = max_total_vaccinations.to_list()
```

```python
from bokeh.plotting import figure, output_file, show
import bokeh.io

p = figure(x_axis_label= 'Total Vaccinations per Capita (By Country)', x_range = [0, 1.5], y_range = countries, tools = 'hover'
          , plot_height = 1500)

p.hbar(y = countries, right = counts, height = 0.6)
p.ygrid.grid_line_color = None

bokeh.io.show(p)
```
